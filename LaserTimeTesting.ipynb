```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
from scipy.optimize import curve_fit

def gaus(x, normalization,location, stdev):
    u = (x - location)/stdev
    return normalization* np.exp(-0.5*u**2)
```

## With T-Junction

```python
with open('Algorithm_Test_Laser_TJunction_New.pkl', 'rb') as f:
    p2ps = pickle.load(f)
```

```python
plt.rcParams["figure.figsize"] = (18,6)
methods = ['fit','smooth','base','raw']
colors=["blue","orange","green","red","purple","cyan", "black"]
for i in range(len(methods)):
    count,binedges,bars=plt.hist(p2ps.eval(f"ch4_time_{methods[i]}-ch1_time_{methods[i]}"),35,range=(25,30),density=True,color = colors[i],histtype="step",label = f"{methods[i]} method")
    bincenters = binedges[1:]/2+binedges[:-1]/2
    coeff, covar = curve_fit(gaus, bincenters, count,p0=(1,29,1))
    plt.plot(np.linspace(coeff[1]-3,coeff[1]+3,500),gaus(np.linspace(coeff[1]-3,coeff[1]+3,500),*coeff),label=f"Gaussian Fit: Width {coeff[2]*1000:.1f} ps",color=colors[i])
plt.legend()
plt.title("With T-Junction Time Difference")
plt.xlabel("Time Difference (ns)")
```

```python
methods = ['fit','smooth','base','raw']
colors=["blue","orange","green","red","purple","cyan", "black"]
for i in range(len(methods)):
    count,binedges,bars=plt.hist(p2ps.eval(f"ch1_time_{methods[i]}"),50,range=(64,69),density=True,color = colors[i],histtype="step",label = f"{methods[i]} method")
    bincenters = binedges[1:]/2+binedges[:-1]/2
plt.legend()
plt.title("Laser With T-Junction")
plt.xlabel("Time of Arrival (ns)")
```
