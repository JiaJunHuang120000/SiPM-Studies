```python
import numpy as np
import matplotlib.pyplot as plt
import csv

# Define function which loads a file titled "textfilename".txt which contains at least two collumns of data: 
    # The first (0th) collumn will become the x-axis values.  The last (-1st) collumn will become the y-axis values
# Output of function is a list containing two lists;
    # The first list is the x-axis values.  The second list is the y-axis values.
def textloader(textfilename):
    graph = [[],[],[],[]]
    data = np.loadtxt(str(textfilename + ".txt"),float)
    graph0 = data[:,0]
    graph1 = data[:,-1]
    
    graph[0] = graph0 #x-axis values
    graph[1] = graph1 #y-axis values
    graph[2] = len(graph[0]) #number of bins/data points
    graph[3] = int(sum(graph[1])) #integral of distribution (if histogram)
    return graph

def csvloader(csvfilename):
    with open(str(csvfilename + ".csv")) as examp:
        reader = csv.reader(examp)
        print(reader)
        read = list(reader)
    
    graph = [[],[],[],[]]
    for i in range(len(read)):
        graph[0].append(read[i][0])
        graph[1].append(read[i][-1])
    graph[2] = len(graph[0])
    graph[3] = int(sum(graph[1]))
    return graph

# Find the index from xvals of the value which is closest to xval.  If more than one exact match, returns the lowest index
# xvals should be spaced less than 66e6 apart
def findindex(xvals,xval): 
    minima = 66e6
    for i,elem in enumerate(xvals):
        # Update minimum difference from xval and index of the minimum difference
        if abs(xval-elem) < minima: 
            minima = abs(xval-elem) 
            index = i
    return index

def histtoplot(histdata,bins): #data to be binned, number of bins
    graph = []
    newbins=[]
    count,bincount,bars=plt.hist(histdata,bins)
    
    for i in range(len(bincount)-1):
        newbins.append(bincount[i]/2+bincount[i+1]/2)
    graph.append(newbins)
    graph.append(count)
    graph.append(len(count))
    graph.append(int(sum(count)))
    return graph #a list of 4 elements: a list of x values, a list of y values, the number of data points, & the integral of the distribution
```
