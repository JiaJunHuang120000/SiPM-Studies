```python
import numpy as np
import matplotlib.pyplot as plt
```

## Load Data

```python
#  !!input document must be delineated with two collumns of (x,y) points!!
#  Graph two functions at different x points on one graph
data1 = np.loadtxt(".txt",float)
data2 = np.loadtxt("sintest.txt",float)

Xaxis1 = data1[:,0]
Yaxis1 = data1[:,1]

Xaxis2 = data2[:,0]
Yaxis2 = 100*data2[:,1]

```

```python
#  !!input document must be delineated with three collumns of (x_min,x_max,y) points!!
#  Graph two histograms with different bins on one graph
hist1 = np.loadtxt("coincidence_ch1.txt",float)
hist2 = np.loadtxt("background_ch1.txt",float)
hist3 = np.loadtxt("bknd_noco_ch1.txt",float)

Xhist1 = hist1[:,0]
Yhist1 = hist1[:,2]

Xhist2 = hist2[:,0]
Yhist2 = hist2[:,2]

Xhist3 = hist3[:,0]
Yhist3 = hist3[:,2]
```

## Make Plots

```python
fig,ax = plt.subplots(1,1,figsize=(13,8))

ax.set_title(title, size=20)
ax.set_xlabel(xlabel, size=20)
ax.set_ylabel(ylabel, size=20)
ax.plot(Xaxis1,Yaxis1,"-")
ax.plot(Xaxis2,Yaxis2,"-")

#ax[1].set_title("Fourier Transform, Zoom", size=20)
#ax[1].set_xlabel("Frequency", size=20)
#ax[1].set_ylabel("Amplitude", size=20)
#ax[1].bar(Xaxis[0:N//2],Yaxis[0:N//2])
```
