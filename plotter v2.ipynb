```python
# Define a function which loads a file titled "textfilename".txt which contains three collumns of data: 
    # (0) bin left bound, (1) bin right bound, (2) count (this is the format of the DRS4 GUI histogram data)
# Output of function is a list containing two lists;
    # The first list is the x-axis values.  The second list is the y-axis values.

def textloader(textfilename):
    graph = [[],[]]
    data = np.loadtxt(str(textfilename + ".txt"),float)
    graph0 = data[:,0]
    graph1 = data[:,2]
    
    graph[0] = graph0
    graph[1] = graph1
    return graph
```
